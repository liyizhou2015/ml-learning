## 1. 导入库

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, validation_curve, learning_curve
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, r2_score

# 设置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

print("库导入成功！")
```

## 2. 生成非线性数据

```python
# 设置随机种子
np.random.seed(42)

# 生成非线性数据: y = 0.5 * x^3 - 2 * x^2 + x + 2 + noise
n_samples = 100
X = np.linspace(-3, 5, n_samples)
y_true = 0.5 * X**3 - 2 * X**2 + X + 2
noise = np.random.normal(0, 3, n_samples)
y = y_true + noise

# 转换为二维数组
X = X.reshape(-1, 1)

# 划分数据集
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# 排序以便绘图
sort_idx_train = np.argsort(X_train.flatten())
X_train_sorted = X_train[sort_idx_train]
y_train_sorted = y_train[sort_idx_train]

sort_idx_test = np.argsort(X_test.flatten())
X_test_sorted = X_test[sort_idx_test]
y_test_sorted = y_test[sort_idx_test]

print(f"训练集大小: {len(X_train)}")
print(f"测试集大小: {len(X_test)}")
```

```python
# 可视化数据
plt.figure(figsize=(10, 6))
plt.scatter(X_train, y_train, alpha=0.6, color='blue', label='训练数据')
plt.scatter(X_test, y_test, alpha=0.6, color='green', label='测试数据')

# 绘制真实函数
X_line = np.linspace(-3, 5, 200).reshape(-1, 1)
y_line = 0.5 * X_line**3 - 2 * X_line**2 + X_line + 2
plt.plot(X_line, y_line, 'r-', linewidth=2, label='真实函数')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('非线性数据集')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## 3. 欠拟合：线性模型

```python
# 训练线性模型（degree=1）
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

# 预测
y_train_pred_linear = linear_model.predict(X_train_sorted)
y_test_pred_linear = linear_model.predict(X_test_sorted)

# 评估
r2_train_linear = r2_score(y_train_sorted, y_train_pred_linear)
r2_test_linear = r2_score(y_test_sorted, y_test_pred_linear)

print("线性模型（欠拟合）:")
print(f"训练集 R²: {r2_train_linear:.4f}")
print(f"测试集 R²: {r2_test_linear:.4f}")
print(f"训练集 RMSE: {np.sqrt(mean_squared_error(y_train_sorted, y_train_pred_linear)):.4f}")
print(f"测试集 RMSE: {np.sqrt(mean_squared_error(y_test_sorted, y_test_pred_linear)):.4f}")
```

```python
# 可视化欠拟合
plt.figure(figsize=(10, 6))
plt.scatter(X_train, y_train, alpha=0.6, color='blue', label='训练数据')
plt.scatter(X_test, y_test, alpha=0.6, color='green', label='测试数据')

# 绘制线性拟合
X_plot = np.linspace(-3, 5, 200).reshape(-1, 1)
y_plot_linear = linear_model.predict(X_plot)
plt.plot(X_plot, y_plot_linear, 'orange', linewidth=2, label='线性拟合 (欠拟合)')

# 绘制真实函数
y_plot_true = 0.5 * X_plot**3 - 2 * X_plot**2 + X_plot + 2
plt.plot(X_plot, y_plot_true, 'r--', linewidth=2, label='真实函数')

plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'欠拟合示例 (R²={r2_test_linear:.3f})')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print("\n欠拟合特征:")
print("- 模型过于简单，无法捕捉数据的非线性模式")
print("- 训练集和测试集表现都不好")
print("- 高偏差 (High Bias)")
```

## 4. 多项式回归

```python
# 创建多项式特征（degree=3）
poly_features = PolynomialFeatures(degree=3, include_bias=False)
X_train_poly = poly_features.fit_transform(X_train)
X_test_poly = poly_features.transform(X_test)

print(f"原始特征形状: {X_train.shape}")
print(f"多项式特征形状: {X_train_poly.shape}")
print(f"\n多项式特征名称: {poly_features.get_feature_names_out(['X'])}")
```

```python
# 训练多项式回归模型
poly_model = LinearRegression()
poly_model.fit(X_train_poly, y_train)

# 预测
y_train_pred_poly = poly_model.predict(poly_features.transform(X_train_sorted))
y_test_pred_poly = poly_model.predict(poly_features.transform(X_test_sorted))

# 评估
r2_train_poly = r2_score(y_train_sorted, y_train_pred_poly)
r2_test_poly = r2_score(y_test_sorted, y_test_pred_poly)

print("多项式回归 (degree=3):")
print(f"训练集 R²: {r2_train_poly:.4f}")
print(f"测试集 R²: {r2_test_poly:.4f}")
print(f"训练集 RMSE: {np.sqrt(mean_squared_error(y_train_sorted, y_train_pred_poly)):.4f}")
print(f"测试集 RMSE: {np.sqrt(mean_squared_error(y_test_sorted, y_test_pred_poly)):.4f}")

print(f"\n模型系数:")
for name, coef in zip(poly_features.get_feature_names_out(['X']), poly_model.coef_):
    print(f"  {name}: {coef:.4f}")
print(f"  截距: {poly_model.intercept_:.4f}")
```

```python
# 可视化多项式拟合
plt.figure(figsize=(10, 6))
plt.scatter(X_train, y_train, alpha=0.6, color='blue', label='训练数据')
plt.scatter(X_test, y_test, alpha=0.6, color='green', label='测试数据')

# 绘制多项式拟合
y_plot_poly = poly_model.predict(poly_features.transform(X_plot))
plt.plot(X_plot, y_plot_poly, 'purple', linewidth=2, label='多项式拟合 (degree=3)')

# 绘制真实函数
plt.plot(X_plot, y_plot_true, 'r--', linewidth=2, label='真实函数')

plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'多项式回归 (R²={r2_test_poly:.3f})')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## 5. 过拟合：高阶多项式

```python
# 创建高阶多项式特征（degree=15）
poly_high = PolynomialFeatures(degree=15, include_bias=False)
X_train_high = poly_high.fit_transform(X_train)
X_test_high = poly_high.transform(X_test)

print(f"高阶多项式特征数: {X_train_high.shape[1]}")

# 训练高阶多项式模型
high_model = LinearRegression()
high_model.fit(X_train_high, y_train)

# 预测
y_train_pred_high = high_model.predict(poly_high.transform(X_train_sorted))
y_test_pred_high = high_model.predict(poly_high.transform(X_test_sorted))

# 评估
r2_train_high = r2_score(y_train_sorted, y_train_pred_high)
r2_test_high = r2_score(y_test_sorted, y_test_pred_high)

print(f"\n高阶多项式回归 (degree=15):")
print(f"训练集 R²: {r2_train_high:.4f}")
print(f"测试集 R²: {r2_test_high:.4f}")
print(f"训练集 RMSE: {np.sqrt(mean_squared_error(y_train_sorted, y_train_pred_high)):.4f}")
print(f"测试集 RMSE: {np.sqrt(mean_squared_error(y_test_sorted, y_test_pred_high)):.4f}")
```

```python
# 可视化过拟合
plt.figure(figsize=(12, 5))

# 子图1: 拟合曲线
plt.subplot(1, 2, 1)
plt.scatter(X_train, y_train, alpha=0.6, color='blue', label='训练数据')
plt.scatter(X_test, y_test, alpha=0.6, color='green', label='测试数据')

y_plot_high = high_model.predict(poly_high.transform(X_plot))
plt.plot(X_plot, y_plot_high, 'orange', linewidth=2, label='高阶多项式 (过拟合)')
plt.plot(X_plot, y_plot_true, 'r--', linewidth=2, label='真实函数')

plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'过拟合示例 (degree=15)')
plt.legend()
plt.grid(True, alpha=0.3)

# 子图2: 对比三种模型
plt.subplot(1, 2, 2)
plt.scatter(X_train, y_train, alpha=0.4, color='blue', label='训练数据')
plt.scatter(X_test, y_test, alpha=0.4, color='green', label='测试数据')

plt.plot(X_plot, y_plot_linear, 'orange', linewidth=2, label='线性 (欠拟合)')
plt.plot(X_plot, y_plot_poly, 'purple', linewidth=2, label='3阶多项式 (合适)')
plt.plot(X_plot, y_plot_high, 'brown', linewidth=2, label='15阶多项式 (过拟合)')
plt.plot(X_plot, y_plot_true, 'r--', linewidth=2, alpha=0.7, label='真实函数')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('三种模型对比')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\n过拟合特征:")
print("- 模型过于复杂，过度拟合训练数据中的噪声")
print("- 训练集表现很好，测试集表现差")
print("- 高方差 (High Variance)")
```

## 6. 模型复杂度与性能关系

```python
# 测试不同阶数的多项式
degrees = range(1, 16)
train_scores = []
test_scores = []

for degree in degrees:
    # 创建多项式特征
    poly = PolynomialFeatures(degree=degree, include_bias=False)
    X_train_d = poly.fit_transform(X_train)
    X_test_d = poly.transform(X_test)
    
    # 训练模型
    model = LinearRegression()
    model.fit(X_train_d, y_train)
    
    # 评估
    train_score = r2_score(y_train, model.predict(X_train_d))
    test_score = r2_score(y_test, model.predict(X_test_d))
    
    train_scores.append(train_score)
    test_scores.append(test_score)

# 可视化
plt.figure(figsize=(10, 6))
plt.plot(degrees, train_scores, 'o-', color='blue', label='训练集 R²')
plt.plot(degrees, test_scores, 'o-', color='green', label='测试集 R²')
plt.axvline(x=3, color='red', linestyle='--', alpha=0.5, label='最优复杂度')
plt.xlabel('多项式阶数')
plt.ylabel('R² Score')
plt.title('模型复杂度 vs 性能')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# 找到最优阶数
best_degree = list(degrees)[np.argmax(test_scores)]
print(f"最优多项式阶数: {best_degree}")
print(f"最优测试集 R²: {max(test_scores):.4f}")
```

## 7. 学习曲线

```python
# 绘制学习曲线
def plot_learning_curve(model, X, y, title):
    train_sizes, train_scores, test_scores = learning_curve(
        model, X, y, cv=5, n_jobs=-1, 
        train_sizes=np.linspace(0.1, 1.0, 10),
        scoring='neg_mean_squared_error'
    )
    
    train_scores_mean = -np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = -np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    
    plt.plot(train_sizes, train_scores_mean, 'o-', color='blue', label='训练误差')
    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1, color='blue')
    plt.plot(train_sizes, test_scores_mean, 'o-', color='green', label='验证误差')
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color='green')
    
    plt.xlabel('训练样本数')
    plt.ylabel('MSE')
    plt.title(title)
    plt.legend()
    plt.grid(True, alpha=0.3)

plt.figure(figsize=(15, 4))

# 欠拟合（线性模型）
plt.subplot(1, 3, 1)
plot_learning_curve(LinearRegression(), X, y, '欠拟合 - 线性模型')

# 合适（3阶多项式）
plt.subplot(1, 3, 2)
poly_pipeline = Pipeline([
    ('poly', PolynomialFeatures(degree=3)),
    ('linear', LinearRegression())
])
plot_learning_curve(poly_pipeline, X, y, '合适 - 3阶多项式')

# 过拟合（15阶多项式）
plt.subplot(1, 3, 3)
high_pipeline = Pipeline([
    ('poly', PolynomialFeatures(degree=15)),
    ('linear', LinearRegression())
])
plot_learning_curve(high_pipeline, X, y, '过拟合 - 15阶多项式')

plt.tight_layout()
plt.show()
```

## 8. 正则化解决过拟合

```python
# 使用 Ridge 回归（L2正则化）解决过拟合
from sklearn.linear_model import Ridge

# 创建 Pipeline
ridge_pipeline = Pipeline([
    ('poly', PolynomialFeatures(degree=15)),
    ('scaler', StandardScaler()),
    ('ridge', Ridge(alpha=1.0))
])

# 训练
ridge_pipeline.fit(X_train, y_train)

# 预测
y_train_pred_ridge = ridge_pipeline.predict(X_train_sorted)
y_test_pred_ridge = ridge_pipeline.predict(X_test_sorted)

# 评估
r2_train_ridge = r2_score(y_train_sorted, y_train_pred_ridge)
r2_test_ridge = r2_score(y_test_sorted, y_test_pred_ridge)

print("Ridge 回归 (degree=15, alpha=1.0):")
print(f"训练集 R²: {r2_train_ridge:.4f}")
print(f"测试集 R²: {r2_test_ridge:.4f}")
print(f"\n对比:")
print(f"无正则化 R²: {r2_test_high:.4f}")
print(f"Ridge R²: {r2_test_ridge:.4f}")
print(f"提升: {r2_test_ridge - r2_test_high:.4f}")
```

```python
# 可视化正则化效果
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(X_train, y_train, alpha=0.4, color='blue', label='训练数据')
plt.scatter(X_test, y_test, alpha=0.4, color='green', label='测试数据')
y_plot_ridge = ridge_pipeline.predict(X_plot)
plt.plot(X_plot, y_plot_ridge, 'purple', linewidth=2, label='Ridge 正则化')
plt.plot(X_plot, y_plot_true, 'r--', linewidth=2, label='真实函数')
plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'Ridge 正则化 (R²={r2_test_ridge:.3f})')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.scatter(X_train, y_train, alpha=0.4, color='blue', label='训练数据')
plt.scatter(X_test, y_test, alpha=0.4, color='green', label='测试数据')
plt.plot(X_plot, y_plot_high, 'brown', linewidth=2, label='无正则化 (过拟合)')
plt.plot(X_plot, y_plot_ridge, 'purple', linewidth=2, label='Ridge 正则化')
plt.plot(X_plot, y_plot_true, 'r--', linewidth=2, label='真实函数')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('正则化效果对比')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 9. 总结与要点

### 偏差-方差权衡

| 问题 | 特征 | 原因 | 解决方案 |
|------|------|------|----------|
| **欠拟合** | 高偏差，训练/测试误差都高 | 模型太简单 | 增加特征复杂度、使用非线性模型 |
| **过拟合** | 高方差，训练误差低测试误差高 | 模型太复杂 | 正则化、减少特征、增加数据 |
| **合适** | 偏差方差平衡，两者误差都低 | - | - |

### 识别方法

1. **学习曲线**: 观察训练误差和验证误差随样本数的变化
2. **验证曲线**: 观察不同超参数下的性能变化
3. **交叉验证**: 评估模型泛化能力

### 解决方案

| 方法 | 适用场景 |
|------|----------|
| 增加模型复杂度 | 欠拟合 |
| 正则化 (Ridge/Lasso) | 过拟合 |
| 特征选择 | 过拟合、维度灾难 |
| 增加训练数据 | 过拟合 |
| 交叉验证 | 模型选择 |

### 最佳实践

1. ✅ 始终划分训练集/验证集/测试集
2. ✅ 使用交叉验证评估模型
3. ✅ 绘制学习曲线诊断问题
4. ✅ 从简单模型开始，逐步增加复杂度
5. ✅ 使用正则化控制模型复杂度
6. ⚠️ 避免在测试集上调参（数据泄露）
